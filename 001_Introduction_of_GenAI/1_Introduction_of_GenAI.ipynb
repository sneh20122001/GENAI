# What is Generative AI ??

- Generative AI is a type of artificial intelligence that can create new content, such as text, images, audio, and video. It works by learning the patterns and relationships in a dataset of human-created content, and then using those patterns to generate new content.

 EX. Text, Images, Audio, Video


## Workflow of GenAI



```
Data Collection and Preparation
                |
          Model Training
                |
        Content Generation
                |
    Evaluation and Refinement
```



##Why are GenAI models required ?

- Content Creation and Augmentation
- Data Augmentation and Simulation
- Personalization and Customization
- Problem Solving and Innovation
- Enhanced Human-Computer Interaction

## Discriminative vs Generative model

- Discriminative models are like experts at identifying and classifying objects, but they can't create new ones.

- Generative models are like artists who can generate new images, but they may not be as good at classifying existing ones.

## What is LLMs ?

- Large Language Models (LLMs) are a type of artificial intelligence (AI) model that are designed to understand and generate human-like text. They are called "large" because they are trained on massive amounts of text data, often billions of words, which allows them to learn the patterns and structures of language in a very comprehensive way.


###How they work:

  - Deep Learning: LLMs are based on deep learning, a type of machine learning that uses artificial neural networks with many layers to analyze data. These networks can learn complex patterns in the data, such as the relationships between words in a sentence.
  - Transformer Architecture: Many LLMs use a specific type of neural network architecture called a "transformer." Transformers are particularly good at processing sequential data like text, as they can consider the context of words in a sentence to understand their meaning.
  - Training: LLMs are trained on massive datasets of text, which can include books, articles, websites, and code. During training, the model learns to predict the next word in a sentence, which helps it to understand the structure and grammar of language.


###What they can do ?

1. Chatbots
2. Question answering
3. Text summarization
4. Language translation
5. Text generation

###Limitations of LLMs:

- Bias: LLMs can inherit biases from the data they are trained on, which can lead to them generating biased or unfair text.
- Hallucinations: LLMs can sometimes generate text that is factually incorrect or nonsensical.
- Computational cost: Training and running LLMs can be very computationally expensive.
