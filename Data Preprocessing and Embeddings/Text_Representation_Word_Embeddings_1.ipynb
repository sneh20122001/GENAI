##Bag of words

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'text':['people watch video',
                           'video watch video',
                           'people write comment',
                           'video write comment'],
                   'output':[1,1,0,0]})
```

## N-grams

N-grams are continuous sequences of words or symbols, or tokens in a document.

```python
df = pd.DataFrame({'text':['people watch video',
                           'video watch video',
                           'people write comment',
                           'video write comment'],
                   'output':[1,1,0,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
```

```python
print(cv.vocabulary_)
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(3,3))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

## TF-IDF (Term frequency- Inverse document frequency)

Term frequency-inverse document frequency (tf-idf) is a measure of how important a word is to a document in a collection of documents.

```python
df = pd.DataFrame({'text':['people watch video',
                           'video watch video',
                           'people write comment',
                           'video write comment'],
                   'output':[1,1,0,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
```

```python
arr = tfidf.fit_transform(df['text']).toarray()
```

```python
arr
```

```python
print(tfidf.idf_)
```
